```python
from transformers import pipeline

classifier = pipeline("image-classification", model='TornikeO/vit-base-patch16-224-imigue')
```

```python
from datasets import load_dataset

data = load_dataset('TornikeO/imigue', split='train[:3000]')
```

```python
from IPython.display import display


for i in range(0, len(data), 100):
    sample = data[i]
    # print(sample)
    sample_img = sample['image']
    sample_img.thumbnail((224, 224))
    sample_class = sample['label'] # Don't forget this!
    pred = classifier(sample_img)
    print('Actual class ',  classifier.model.config.id2label[sample_class], 'predicted', pred[0]['label'], f' {pred[0]["score"] * 100:.3f} % ')
    display(sample_img)
```
